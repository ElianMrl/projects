## Alzheimer Dataframe

### 3. Load Alzheimer Dataframe & Basic EDA 

```python
# https://www.kaggle.com/datasets/ananthu19/alzheimer-disease-and-healthy-aging-data-in-us
df = pd.read_csv("Alzheimer.csv")
```

### 4. Label Encoding 

```python
ch = preprocessing.LabelEncoder()

df['LocationAbbr']= ch.fit_transform(df['LocationAbbr'])
df['LocationDesc']= ch.fit_transform(df['LocationDesc'])
df['Class']= ch.fit_transform(df['Class'])
df['Topic']= ch.fit_transform(df['Topic'])
df['Question']= ch.fit_transform(df['Question'])
df['StratificationCategory1']= ch.fit_transform(df['StratificationCategory1'])
df['Stratification1']= ch.fit_transform(df['Stratification1'])
df['StratificationCategory2']= ch.fit_transform(df['StratificationCategory2'])
df['Stratification2']= ch.fit_transform(df['Stratification2'])
df['Geolocation']= ch.fit_transform(df['Geolocation'])
```

```python
df.head()
```

```python
df
```

### 5. Data Reduction 

```python
np.random.seed(15)

# the Random Forest can not handle large amount of data, therefore, I took a random sample of 5% of the data 
df = df.sample(frac=0.05) 
df
```

#### Dividing Class variable from attributes 

```python
# Class - Note: there are multiple labels 'Overall Health', 'Mental Health', 'Smoking and Alcohol Use',etc.. 
y = df['Class']

# Attributes of the objects 
X = df.drop(['Class'], axis = 1)
X.head()
```

### 6. Algorithm Generation & Train Test Split

```python
# This function fit and score models
def fit_and_score(models,      # insert models and models' names: {"Random Forest": RandomForestClassifier()}
                  X,           # attributes of data set
                  y):          # class of data set 
    
    # Create a KFold object with 10 folds
    kf = KFold(n_splits=10, shuffle=True, random_state=15)
    
    # keep model scores
    model_scores = {}
    
    # Loop through models
    for name, model in models.items():
        
        # start recording time
        start = time.time()
        
        # Compute the cross-validation score
        scores = cross_val_score(model, X, y, cv=kf)
        
        # Evaluate the model and append its score to model_scores
        model_scores[name] = np.mean(scores)
        
        end = time.time()
        total_time = end - start
        
        # Evaluation and Test Results: 
        print("Algorithm: ", name)
        print("Cross-validation scores: ", scores)
        print("Average score: ", np.mean(scores))
        print("Standard deviation: ", np.std(scores))
        print("Time: ", total_time, '\n')
        
    return (model_scores)
```

### 7. Algorithm Evaluation

```python
# For the sake of following the lecture examples, I will used the gini criterion, the number of features be equal to the 
# square root of features for when splitting the nodes and as mentioned in the lecture notes, I am using the number of 
# estimators to be equal to the number of rows of the data set. But if I ever want to make some changes, I can always 
# made as many changes as I desired.

models = {"Multinomial Naive Bayes": MultinomialNB(),
          "Random Forest": RandomForestClassifier(n_estimators=len(y), criterion='gini', max_features='sqrt')}

model_scores = fit_and_score(models=models,
                             X=X,
                             y=y)
model_scores

print("the lenght of the data: ",len(y))
```

```python
plt.subplots(figsize =(8, 6))

modelos = list(model_scores.keys())
accuracy = list(model_scores.values())

plt.bar(range(len(model_scores)), accuracy, tick_label=modelos, color=['blue', 'orange', 'green'])

ax = plt.gca()
plt.bar_label(ax.containers[0])

plt.title('Accuracy of Each Model', fontweight ='bold')

plt.xlabel('Models', fontweight ='bold', fontsize = 12)
plt.ylabel('Accuracy', fontweight ='bold', fontsize = 12)

plt.show()
```

## Babies Dataframe

### 7. Load Babies Dataframe & Basic EDA

```python
# https://www.kaggle.com/datasets/debjeetdas/babies-birth-weight
df = pd.read_csv("babies.csv")
df
```

```python
df = df.drop(columns=['case'])
```

```python
df['bwt'].unique()
```

```python
# change birthweight in ounces to grams
df.loc[:,'bwt'] *= 28.3495
```

```python
df['bwt'].unique()
```

```python
max(df['bwt'])
```

```python
# A Normal Birth Weight is between 2500-4000 grams
# A Low Birth Weight is between 1500-2500 grams
# A Very Low Birth Weight is less than 1500 grams
# A 

# citation: Eghbalian, Fatemeh & Monsef, A. & Taghdiri, M.M.. (2008). Neonatal seizures; Etiology and frequency. 
# Iranian Journal of Child Neurology. 2. 
# url: https://www.researchgate.net/figure/Birth-weights-of-the-neonates-with-seizure_tbl1_279640858

df['bwt_label'] = pd.cut(x=df['bwt'], bins=[0, 1500, 2500, 4000, 6000],
                     labels=['Very Low', 'Low', 'Normal', 'Heavy'])
    
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

### 8. Label Encoding

```python
ch = preprocessing.LabelEncoder()

df['bwt_label']= ch.fit_transform(df['bwt_label'])
df = df.drop(columns=['bwt'])
```

```python
df.head()
```

#### Dividing Class variable from attributes

```python
# Class - Note: there are multiple labels 'Very Low', 'Low', 'Normal', 'Heavy' 
# The goal is to predict the babies birth way base on the mother's conditions
y = df['bwt_label']

# Attributes of the objects 
X = df.drop(['bwt_label'], axis = 1)
X.head()
```

### 9. Algorithm Generation & Evaluation

```python
models = {"Multinomial Naive Bayes": MultinomialNB(),
          "Random Forest": RandomForestClassifier(n_estimators=len(y), criterion='gini', max_features='sqrt')}

model_scores = fit_and_score(models=models,
                             X=X,
                             y=y)
model_scores

print("the lenght of the data: ",len(y))
```

```python
plt.subplots(figsize =(8, 6))

modelos = list(model_scores.keys())
accuracy = list(model_scores.values())

plt.bar(range(len(model_scores)), accuracy, tick_label=modelos, color=['blue', 'orange', 'green'])

ax = plt.gca()
plt.bar_label(ax.containers[0])

plt.title('Accuracy of Each Model', fontweight ='bold')

plt.xlabel('Models', fontweight ='bold', fontsize = 12)
plt.ylabel('Accuracy', fontweight ='bold', fontsize = 12)

plt.show()
```
