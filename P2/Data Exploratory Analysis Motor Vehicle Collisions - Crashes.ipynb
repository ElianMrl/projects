Database: [Motor Vehicle Collisions - Crashes](https://data.cityofnewyork.us/Public-Safety/Motor-Vehicle-Collisions-Crashes/h9gi-nx95)

The Motor Vehicle Collisions crash table contains details on the crash event. Each row represents a crash event. The Motor Vehicle Collisions data tables contain information from all police reported motor vehicle collisions in NYC. The police report (MV104-AN) is required to be filled out for collisions where someone is injured or killed, or where there is at least $1000 worth of damage. It should be noted that the data is preliminary and subject to change when the MV-104AN forms are amended based on revised crash details.For the most accurate, up to date statistics on traffic fatalities, please refer to the NYPD Motor Vehicle Collisions page (updated weekly) or Vision Zero View (updated monthly).

### Create a Test Set 

Large datasets can be cumbersome to process, especially on standard personal computers. A well-chosen representative subset can significantly speed up computations, making preliminary analyses faster and more interactive.

By working with a smaller, but representative subset, you can better understand the distributions, relationships, and potential anomalies in your data without getting overwhelmed.

Not all systems have the capacity to handle large datasets. By sampling a representative subset, you can still perform meaningful analyses without the need for specialized big data tools or infrastructure.

A **stratified sample** is a sampling method used to produce a sample that is representative of the population across certain categories or groups.

### Removing Outliers 

```python
from sklearn.cluster import DBSCAN
from sklearn.ensemble import RandomForestClassifier
```

```python
bronx = df[df["BOROUGH"] == 'BRONX'].copy()
brooklyn = df[df["BOROUGH"] == 'BROOKLYN'].copy()
manhattan = df[df["BOROUGH"] == 'MANHATTAN'].copy()
queens = df[df["BOROUGH"] == 'QUEENS'].copy()
statenIs = df[df["BOROUGH"] == 'STATEN ISLAND'].copy()
```

```python
def rmNoise(borought, noise='rm', fr=0.5):
    data = borought.sample(frac=fr, random_state=42)
    coords = data[['LATITUDE', 'LONGITUDE']].values
    db = DBSCAN(eps=0.002, min_samples=50).fit(coords)  # you might need to adjust eps and min_samples
    data['labels'] = db.labels_
     
    if noise == 'rm':
        df_filtered = data[data['labels'] != -1]  # -1 indicates noise in DBSCAN
        return(df_filtered)

    else:
        data['is_outlier'] = (data['labels'] == -1).astype(int)    
        return(data)
```

```python
def rmOutliers(newData, fullData):
    X_train = newData[['LATITUDE', 'LONGITUDE']]
    y_train = newData['is_outlier']
    
    clf = RandomForestClassifier(random_state=42)
    clf.fit(X_train, y_train)

    new_data_predictions = clf.predict(fullData[['LATITUDE', 'LONGITUDE']])
    
    fullData['is_outlier'] = new_data_predictions

    return(fullData[fullData['is_outlier'] == 0])
```

```python
newBrooklyn = rmNoise(brooklyn, noise='data', fr=0.3)
```

```python
brooklynClean = rmOutliers(newBrooklyn, brooklyn)
```
